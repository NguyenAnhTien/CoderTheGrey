```python
import time
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (30, 20)
plt.rcParams.update({'font.size': 30})
```

```python
def is_prime_o_sqrt_n(number):
    if number <= 1:
        return False
    for factor in range(2, int(number ** 0.5) + 1):
        if number % factor == 0:
            return False
    return True

def is_prime_o_n(number):
    if number <= 1:
        return False
    for factor in range(2, number):
        if number % factor == 0:
            return False
    return True
```

```python
START = 50000
times1 = []
times2 = []
for number in range(START, START * 2):
    start = time.time()
    is_prime_o_n(number)
    end = time.time()
    length = end - start
    times1.append(length)
for number in range(START, START * 2):
    start = time.time()
    is_prime_o_sqrt_n(number)
    end = time.time()
    length = end - start
    times2.append(length)
plt.plot(list(range(START, START * 2)), times1, '-g', linewidth=1.5)
plt.plot(list(range(START, START * 2)), times2, '-r', linewidth=1.5)
```

```python
def is_prime_o_sqrt_n(number):
    num_ops = 0
    if number <= 1:
        num_ops += 1
        return num_ops, False
    for factor in range(2, int(number ** 0.5) + 1):
        num_ops += 1
        if number % factor == 0:
            return num_ops, False
    return num_ops, True

def is_prime_o_n(number):
    num_ops = 0
    if number <= 1:
        num_ops += 1
        return num_ops, False
    for factor in range(2, number):
        num_ops += 1
        if number % factor == 0:
            return num_ops, False
    return num_ops, True

START = 50000
times1 = []
times2 = []
for number in range(START, START * 2):
    num_ops, _ = is_prime_o_n(number)
    times1.append(num_ops)
for number in range(START, START * 2):
    num_ops, _ = is_prime_o_sqrt_n(number)
    times2.append(num_ops)
plt.plot(list(range(START, START * 2)), times1, '-g')
plt.plot(list(range(START, START * 2)), times2, '-r')
```
